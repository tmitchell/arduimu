```python
from csv import DictReader
from dateutil import parser
import simplekml
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
import numpy as np
```

```python
FILENAME = 'LOG00128.clean.csv'
```

```python
def easy_scatter(data, x_field, y_field):
    plt.scatter(data[x_field], data[y_field])
    ax = plt.gca()
    ax.set_xlabel(x_field)
    ax.set_ylabel(y_field)
    plt.show()
```

```python
# data = import_data(FILENAME)
with open(FILENAME, 'rb') as f:
    headers = f.readline().strip().split(',')
# load the date as an object not float
dd = [(headers[0], 'object')] + [(a, 'd') for a in headers[1:]]
conv = {0: lambda s: parser.parse(s)}
data = np.genfromtxt(FILENAME, delimiter=',', names=True, converters=conv, dtype=dd)
```

```python
easy_scatter(data, 'LON', 'LAT')
```

```python
easy_scatter(data, 'MGH', 'YAW')
```

```python
easy_scatter(data, 'COG_180', 'MGH')
```

```python
easy_scatter(data, 'SEC', 'ALT')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['LON'], data['LAT'], data['ALT'])
ax.set_xlabel('LON')
ax.set_ylabel('LAT')
ax.set_zlabel('ALT')

plt.show()
```

```python
kml = simplekml.Kml()
for lat,lon,alt,utc in zip(data['LON'], data['LAT'], data['ALT'], data['UTC']):
    if not np.isnan(lat) and not np.isnan(lon):        
        p = kml.newpoint(coords=[(lat, lon, alt)])
        p.timestamp.when = utc
        p.style.iconstyle.scale = 0.25
kml.save('test.kml')
```
